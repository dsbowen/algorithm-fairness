```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(style='white', font_scale=1.15)
```

```python
x = np.linspace(0, 1)
y = 1-x**2
```

```python
def style(ax):
    ax.set(xlabel='Better performance  -->', ylabel='More interpretable  -->', ylim=(-.1, 1.5), yticks=[], xticks=[])
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    return ax

ax = style(sns.lineplot(x=x, y=y))
ax.figure.savefig('figures/perf_interpret_0.png')
ax
```

```python
df = pd.DataFrame({
    'Performance': list(x)*2,
    'Interpretability': list(1-x**2) + list(1.2-x**2),
    'Technique': ['None']*x.shape[0] + ['SHAP']*x.shape[0]
})
ax = style(sns.lineplot(data=df, x='Performance', y='Interpretability', hue='Technique'))
ax.figure.savefig('figures/perf_interpret_1.png', bbox_inches='tight')
ax
```

```python
df = pd.DataFrame({
    'Performance': list(x)*3,
    'Interpretability': list(1-x**2) + list(1.2-x**2) + list(1.4-x**2),
    'Technique': ['None']*x.shape[0] + ['SHAP']*x.shape[0] + ['G-SHAP']*x.shape[0]
})
ax = style(sns.lineplot(data=df, x='Performance', y='Interpretability', hue='Technique'))
ax.figure.savefig('figures/perf_interpret_2.png', bbox_inches='tight')
ax
```

```python
df = pd.DataFrame({
    'Original risk score': list(x)*3,
    'Adjusted risk score': list(x) + list(x**1.14) + list(2*x-x**1.14),
    'Race': ['45 degree']*x.shape[0] + ['Black']*x.shape[0] + ['White']*x.shape[0]
})
ax = sns.lineplot(data=df, x='Original risk score', y='Adjusted risk score', hue='Race', palette='gist_earth')
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.figure.savefig('figures/adjustment.png', bbox_inches='tight')
ax
```
